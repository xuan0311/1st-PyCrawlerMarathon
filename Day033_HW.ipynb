```python
'''Your Code'''

#載入套件
import requests
from bs4 import BeautifulSoup

# 使用 GET 方式取得網頁資訊
res = requests.get('https://github.com/new', verify = False)

'''確認從伺服器傳回的狀態碼'''
# 伺服器回應的狀態碼
print("伺服器回應的狀態碼:",res.status_code)#如果顯示 200 就代表沒問題

# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(res.text, 'html.parser') #指定html.parser作為解析器
print(soup.text)
```
