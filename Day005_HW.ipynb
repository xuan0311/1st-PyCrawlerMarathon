1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true') 
response = r.text  #模擬發送請求

response
```

```python
import json
data = json.loads(response)  #模擬發送請求
count = 0
for d in data:
    print(d['id'])
    count +=1

print("總共傳回", count,"筆資料")

print("\n每一筆資料包含欄位:",data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
import json
data = json.loads(response)  #模擬發送請求
for d in data:
    print(d['title'],d['createdAt'],d['commentCount'], d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
#popular
import requests
import json
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true') 
response = r.text  #模擬發送請求
data = json.loads(response)  #模擬發送請求
pop_count = 0
pop_comment = 0
pop_like = 0
for d in data:
    pop_count =pop_count + 1
    pop_comment = pop_comment + d['commentCount']
    pop_like = pop_like + d['likeCount']
avg_pop_comment = pop_comment / pop_count
avg_pop_like = pop_like / pop_count

#unpopular
import requests
import json
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=false') 
response = r.text 
data = json.loads(response)  #模擬發送請求
unpop_count = 0
unpop_comment = 0
unpop_like = 0
for d in data:
    unpop_count = unpop_count + 1
    unpop_comment = unpop_comment + d['commentCount']
    unpop_like = unpop_like + d['likeCount']
avg_unpop_commment = unpop_comment / unpop_count
avg_unpop_like =  unpop_like / unpop_count    

#result
print('熱門文章的「平均留言人數」', avg_pop_comment ,'與「平均按讚人數」:', avg_pop_like)
print('非熱門文章的「平均留言人數」',avg_unpop_commment,'與「平均按讚人數」',avg_unpop_like)
```
