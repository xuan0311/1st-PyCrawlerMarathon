## 根據範例，取出今天所有的新聞

## 打開瀏覽器

```python
import time
today = time.strftime("%Y/%m/%d", time.localtime())
print(today)
```

```python
'''
Your Code
'''
#載入套件
from selenium import webdriver
from bs4 import BeautifulSoup
#使用webdriver.Chrime()建立Google Chrime瀏覽器物件
browser = webdriver.Chrome(executable_path='chromedriver')
#以get()方法連結到指定網址
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 每個兩秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(2) #是Python程序暫停執行2秒
    #通過左邊控制橫向和縱向滾動條scrollTo(x, y）
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source # browser.page_source網頁原始碼
# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(html_source, "html5lib") #指定html.parser作為解析器

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text.split(' ')[0] == today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```

# 取出現在時間兩小時內的新聞

```python
#載入套件
from datetime import datetime
from datetime import timedelta
from selenium import webdriver
from bs4 import BeautifulSoup
#使用webdriver.Chrime()建立Google Chrime瀏覽器物件
browser = webdriver.Chrome(executable_path='chromedriver')
#以get()方法連結到指定網址
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2) #是Python程序暫停執行2秒
    #通過左邊控制橫向和縱向滾動條scrollTo(x, y）
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source # browser.page_source網頁原始碼
# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(html_source, "html5lib") #指定html.parser作為解析器

now = datetime.now()

for d in soup.find(class_="part_list_2").find_all('h3'):
    t = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")  #strptime 將 datetime string 轉換成object
    if (now - t <= timedelta(hours=2)):
        print(t, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```

## 取出今天下午三點到五點的新聞

```python
'''
Your Code
'''
#載入套件
from datetime import datetime
from datetime import timedelta
from selenium import webdriver
from bs4 import BeautifulSoup
#使用webdriver.Chrime()建立Google Chrime瀏覽器物件
browser = webdriver.Chrome(executable_path='chromedriver')
#以get()方法連結到指定網址
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2) #是Python程序暫停執行2秒
    #通過左邊控制橫向和縱向滾動條scrollTo(x, y）
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source # browser.page_source網頁原始碼
# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(html_source, "html5lib") #指定html.parser作為解析器

for d in soup.find(class_="part_list_2").find_all('h3'):
    t = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M") #strptime 將 datetime string 轉換成object
    if t.hour >= 15 and t.hour <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
#載入套件
from datetime import datetime
from datetime import timedelta
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
#使用webdriver.Chrime()建立Google Chrime瀏覽器物件
browser = webdriver.Chrome(executable_path='chromedriver')
#以get()方法連結到指定網址
browser.get("https://www.ettoday.net/news/news-list.htm")

# 選到三天前的日期，按下查詢按鈕
#以id查詢符合的元素
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2020')
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('1')
selectDay = Select(browser.find_element_by_id('selD'))
selectDay.select_by_value('4')

# Click Button
browser.find_element_by_id('button').click()

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2) #是Python程序暫停執行2秒
    #通過左邊控制橫向和縱向滾動條scrollTo(x, y）
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source # browser.page_source網頁原始碼
# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(html_source, "html5lib") #指定html.parser作為解析器

timeStart = '15:00'
timeEnd   = '17:00'
for d in soup.find(class_="part_list_2").find_all('h3'):
    timeData = d.find(class_="date").text
    #從時間開始抓起
    timeData = timeData[11:len(timeData)]
    if int(timeData[:2]) < int(timeStart[:2]):
        break
    if int(timeData[:2]) < int(timeEnd[:2]) or timeData == '17:00':    
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```
