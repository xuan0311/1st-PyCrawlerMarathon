### ① 印出最新文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
# 引入函式庫
import requests
from bs4 import BeautifulSoup
import pandas as pd
# 想要爬資料的目標網址
url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
# 取回請求的回覆
soup = BeautifulSoup(r.text, "html5lib")
#soup.find(class_="title") 找出被 <div class="part_list_2"> </div>標籤所包圍的區塊
#find_all:找出所有特定的 HTML 標籤節點。
df = pd.DataFrame(columns=['作者', '標題','時間'])
i = 0
for d in soup.find_all(class_="title"):
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        data=r.find_all(class_='article-meta-value')
        df.loc[i] = [data[0].text, data[2].text, data[3].text]
        i+=1
    except:
        continue
df.head(1)
```
