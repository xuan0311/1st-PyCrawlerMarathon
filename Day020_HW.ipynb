```python
#載入套件
from bs4 import BeautifulSoup

# 取回請求的回覆
# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(r.text, 'html.parser') #指定html.parser作為解析器
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
'''
Your Code
'''
d = {}
for tr in table.find_all('tr')[1:]:

    for i, td in enumerate(tr.find_all('td')):
        
        if len(tr.find_all('td')) == 5:
            if i == 0:
                obs = td.text
                d.setdefault(obs, {}) # setdefault() 返回的鍵如果不在字典中，會添加鍵（更新字典）
            if i == 2:
                date = td.text
            if i == 3:
                value = td.text
                d[obs][date] = value
        
        if len(tr.find_all('td')) == 3:
            if i == 0:
                date = td.text
            if i == 1:
                value = td.text
                d[obs][date] = value
print(d)
```
