### ① 取出今天所有的發文

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
# 引入函式庫
import requests
from bs4 import BeautifulSoup
import pandas as pd
# 想要爬資料的目標網址
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
# 取回請求的回覆
soup = BeautifulSoup(r.text, "html5lib")
df = pd.DataFrame(columns=['時間', '分類', '標題'])
#soup.find(class_="part_list_2") 找出被 <div class="part_list_2"> </div>標籤所包圍的區塊
#find_all:找出所有特定的 HTML 標籤節點。
i = 0
for d in soup.find(class_="part_list_2").find_all('h3'): 
    df.loc[i] = [d.find(class_="date").text, d.find_all('em')[-1].text,d.find_all('a')[-1].text]
    i += 1
df.head(50)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''

category = df.groupby('分類').size()
category
```
