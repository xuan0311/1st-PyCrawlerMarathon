```python
'''Your Code'''
#載入套件
import requests
import time
from bs4 import BeautifulSoup

startTime = time.time() # time.time(): 傳回從 1970/1/1 00:00:00 算起至今的秒數
# 使用 GET 方式取得網頁資訊
res = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html')
# 伺服器回應的狀態碼
print("伺服器回應的狀態碼:",res.status_code)#如果顯示 200 就代表沒問題

# 以 BeautifulSoup 解析 HTML 原始碼
soup = BeautifulSoup(res.text, 'html.parser') #指定html.parser作為解析器
print(soup.text)
finishTime = time.time() # time.time(): 傳回從 1970/1/1 00:00:00 算起至今的秒數
print("正常情況的爬蟲所需時間:",finishTime - startTime) # 正常情況的爬蟲所需時間
```
